### Задание: Реализация системы аутентификации и ролей для защищенного доступа

**Описание задачи:**
Создайте Django-приложение для системы управления проектами, в котором пользователи могут быть зарегистрированы в одной из ролей: "Администратор", "Менеджер" или "Пользователь". В зависимости от роли пользователя ограничьте доступ к определенным действиям в приложении.

**Требования к заданию:**

1. **Регистрация и аутентификация**:
   - Настройте регистрацию пользователей, где при создании учетной записи они получают одну из ролей.
   - Реализуйте механизм аутентификации с использованием сессий.
   - Ограничьте доступ к закрытым страницам: неавторизованные пользователи должны перенаправляться на страницу входа.

2. **Ролевые права доступа**:
   - **Администратор**:
     - Имеет полный доступ ко всем действиям, включая создание, обновление и удаление проектов, а также управление пользователями и их ролями.
   - **Менеджер**:
     - Может просматривать проекты, добавлять и редактировать задачи, назначать задачи пользователям, но не может удалять проекты или управлять пользователями.
   - **Пользователь**:
     - Может просматривать проекты, но может добавлять комментарии и отмечать задачи как завершенные.

3. **Доступ к ресурсам**:
   - Настройте декораторы для ограничения доступа к определенным представлениям в зависимости от роли пользователя.
   - Реализуйте специализированные представления для различных типов пользователей с использованием `LoginRequiredMixin` и кастомных прав доступа.
   
4. **Система токенов для API** (опционально, для сложных групп):
   - Реализуйте API-эндпоинты для взаимодействия с проектами и задачами, используя аутентификацию на основе токенов. 
   - Пример: Администратор и менеджер могут создавать проекты и задачи через API, а пользователь — только просматривать их.

5. **Тестирование доступа и логики**:
   - Протестируйте работу приложения в Postman или другом инструменте, убедившись, что пользователи с разными ролями имеют соответствующие права доступа.
   - Напишите тесты для проверки ролей и правил доступа для всех пользователей.

**Ожидаемые результаты:**
- В приложении должна быть реализована система ролей с правильной логикой доступа для каждой роли.
- Доступ к защищенным страницам должен быть ограничен, и пользователи должны иметь возможность видеть только те данные и действия, которые разрешены их ролью.
- API для взаимодействия с проектами и задачами должен быть защищен токенами и иметь аналогичное разграничение прав доступа.

**Проверка выполнения задания:**
Рассматривается корректность реализации ролевых ограничений, соответствие результатов ожиданиям для каждой роли и полнота реализации системы аутентификации и токенов для API (если реализовано).
